```python
import requests
from bs4 import BeautifulSoup
```

```python
for i in range(10):
    url='https://www.flipkart.com/vivo-v15-aqua-blue-64-gb/product-reviews/itmfgzghnmpchy6y?pid=MOBFFH45GVCDABHJ&page=' + str(i)
    r=requests.get(url)
    soup=BeautifulSoup(r.content,'html.parser')
    div_tag=soup.findAll('div',attrs={'id':'container'})[0]
    main_div=div_tag.findAll('p',attrs={'class':'_2xg6Ul'})
    for reviw in main_div:
        print(reviw.text)
```
