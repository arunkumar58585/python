```python
import pandas as pd
df=pd.read_csv("C:\ML\Pandas\weather_data.csv",parse_dates=["day"])
df.set_index("day",inplace=True)
df
```

```python
new_df=df.fillna({'temperature':df.temperature.mean(),'windspeed':df.windspeed.mean(),'event':'No Event'})
new_df
```

```python
new_df=df.fillna(method='ffill')
new_df
```

```python
new_df=df.fillna(method='bfill')
new_df
```

```python
new_df=df.fillna(method='ffill',axis='columns')
new_df
```

```python
new_df=df.fillna(method='ffill',axis='rows',limit=2)
new_df
```

```python
new_df=df.interpolate(method='time')
new_df
```

```python
new_df=df.dropna()
new_df
```

```python
new_df=df.dropna(how='all')
new_df
```

```python
new_df=df.dropna(thresh=1)
new_df
```

```python
dt=pd.date_range("01-01-2017","01-11-2017")
indx=pd.DatetimeIndex(dt)
new_df=df.reindex(indx)
new_df
```
